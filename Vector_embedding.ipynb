```python
import gensim.downloader as api
model = api.load("word2vec-google-news-300") # Downloads the model and return it as an object
```

```python
word_vector = model

print(word_vector['computer']) # Vector of computer word
```

```python
print(word_vector['cats'].shape)
```

## Similar words
### King + Women - Men = ?

```python
# Example of using most_similarity
print(word_vector.most_similar(positive=['king','woman'], negative=['man'], topn=10))
```

```python
print(word_vector.most_similar(positive=['father','woman'], negative=['man'], topn=10))
```

## Similarity score

```python
print(word_vector.similarity('woman','man'))
print(word_vector.similarity('king','queen'))
print(word_vector.similarity('uncle','aunt'))
print(word_vector.similarity('boy','girl'))
print(word_vector.similarity('water','rock'))
```
